```python
# Define the derivative of the function f(x), for gradient descent
def derivative_f(x):
    return x / np.sqrt(x**2 + 5)

# write a function to run the Gradient Descent Algorithm, take input arguments x_0, step_size, and num_iterations; return a list/array of x_k values.
def perform_gradient_descent(initial_x, learning_rate, iterations):
    x_values = [initial_x]
    for _ in range(iterations):
        initial_x = initial_x - learning_rate * derivative_f(initial_x)
        x_values.append(initial_x)
    return x_values

# set up/assign initial values
max_iterations = 50
learning_rate = 1
initial_guess = 6/5+1

# Run gradient descent by calling the function
xs_values = perform_gradient_descent(initial_guess, learning_rate, max_iterations)

# report final value of x by printing out the value
final_value_of_x = xs_values[-1]
print(f"Final value of x: {final_value_of_x}")

# Plot the f(x_k) values vs. iteration number
values_of_f_at_sequence_points = [f(x) for x in xs_values]

plt.plot(range(max_iterations+1), values_of_f_at_sequence_points)
plt.xlabel('k')
plt.ylabel('f(x_k)')
plt.title('$f(x_{k})$ vs. Iteration number')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(84)   # change the seed to be the last two digits of your SB ID.
X = 2 * np.random.rand(100, 2)  # Generate 100 random 2D data points
y = 4 + 4 * X[:, 0] + 0.5 * X[:, 1] + np.random.randn(100)  # Generate y values with noise

# Add a column of ones to X for the bias terms (intercept) in the normal equation
X_b = np.c_[np.ones((100,1)), X]


# Find theta using the normal equation
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)


# Print the intercept and coefficients of the regression plane
print("Intercept: ", theta_best[0])
print("Coefficients: ", theta_best[1:])

# Plot the generated data and the regression plane
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[:,0], X[:,1], y)

x0_range = np.linspace(X[:,0].min(),X[:,0].max(), num=100)
x1_range = np.linspace(X[:,1].min(),X[:,1].max(), num=100)
x0_values, x1_values = np.meshgrid(x0_range,x1_range)
y_pred_surface = theta_best[0] + theta_best[1]*x0_values + theta_best[2]*x1_values

ax.plot_surface(x0_values,x1_values,y_pred_surface,alpha=0.5,cmap='viridis')
plt.title('Linear Regression Plane')
ax.scatter(X[:,0], X[:,1], y, label='Original Data', color='blue')
ax.legend()
plt.show()
```

## 4. Logistic Regression with Numpy and Graphing with Matplotlib (20 pts)

In class, we implemented a logistic regression model to perform binary classification. Here, I provided the exact same code that we discussed in class, but the values for learning_rate and num_epochs are modified. I also removed all the comments. You task is to

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function for use in the logistic regression model
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Set the random seed for reproducibility
np.random.seed(595)
# Generate synthetic data for training and testing
X = np.random.rand(2, 100)
w_true = np.array([1.5, -2.5])
b_true = 1.0  # True bias
probabilities = sigmoid(np.dot(w_true.T, X) + b_true)
Y = (probabilities > 0.5).astype(int)
X = X + 0.3 * np.random.rand(2, 100) - 0.1 * np.random.rand(2, 100)

# Split the data into training and testing sets
X_train, X_test = X[:,:80], X[:,80:]
Y_train, Y_test = Y[:80], Y[80:]

# Initialize parameters w and b to zero
w = np.zeros(X.shape[0])
b = 0.0

# Set learning rate and number of epochs for gradient descent optimization 
learning_rate = 0.05  #modified
num_epochs = 1000  #fine-tuned

# Training using Gradient Descent Optimization 
for epoch in range(num_epochs):
    A_train = sigmoid(np.dot(w.T, X_train) + b)

    gradient_with_respect_to_weights = np.dot(X_train, (A_train - Y_train).T) / len(Y_train)
    gradient_with_respect_to_bias = np.mean(A_train - Y_train)

    # Update weights and bias using the gradients computed above
    w -= learning_rate * gradient_with_respect_to_weights
    b -= learning_rate * gradient_with_respect_to_bias


A_train = sigmoid(np.dot(w.T, X_train) + b)
predictions_train = (A_train > 0.5).astype(int)

A_test = sigmoid(np.dot(w.T, X_test) + b)
predictions_test = (A_test > 0.5).astype(int)

train_accuracy = np.mean(predictions_train == Y_train)
test_accuracy = np.mean(predictions_test == Y_test)

print(f"Training Set Accuracy: {train_accuracy:.2f}")
print(f"Test Set Accuracy: {test_accuracy:.2f}")
```

```python
# Plot the training data and the decision boundary
plt.scatter(X[0, :80], X[1,:80], c=Y[:80], cmap=plt.cm.Paired)
plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)

# Get current axes and their limits
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# Create a grid to evaluate model
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50), np.linspace(ylim[0], ylim[1], 50))

# Compute sigmoid function over the grid
Z = np.dot(w.T, np.c_[xx.ravel(), yy.ravel()].T) + b
Z = sigmoid(Z)
Z = Z.reshape(xx.shape)

# Draw contour at probability 0.5 (decision boundary)
plt.contour(xx, yy, Z, colors='k', levels=[0.5], alpha=0.5, linestyles=['--'])

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression Decision Boundary')
plt.show()
```
